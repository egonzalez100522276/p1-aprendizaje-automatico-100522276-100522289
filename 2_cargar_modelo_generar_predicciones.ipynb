```python
# Rutas
RUTA_MODELO_FINAL = "/content/p1-aprendizaje-automatico-100522276-100522289/modelo_final.joblib"
RUTA_DATOS_COMPETICION = "/content/p1-aprendizaje-automatico-100522276-100522289/bank_ALL/bank_competition.pkl"

```

## Predicciones sobre el conjunto de competición
La práctica pide utilizar el modelo final para generar predicciones sobre el conjunto de competición, que no contiene la variable objetivo. Estas predicciones se guardarán en un archivo CSV llamado *predicciones.csv*.

```python
import pandas as pd
import numpy as np
import joblib

# Cargar el modelo final
# Nota: En tus variables aparece como .joblib, usaremos esa ruta
pipeline_modelo_final = joblib.load(RUTA_MODELO_FINAL)

# Cargar datos de competición
df_comp = pd.read_pickle(RUTA_DATOS_COMPETICION)

print(f"Modelo cargado desde: {RUTA_MODELO_FINAL}")
print(f"Shape competición: {df_comp.shape}")
display(df_comp.head())
```

```python
# Preprocesamiento de los datos de competición (siguiendo tu lógica)
df_comp_proc = df_comp.copy()

if "pdays" in df_comp_proc.columns:
    df_comp_proc["pdays_no_contacto"] = (df_comp_proc["pdays"] == -1).astype(int)
    df_comp_proc["pdays"] = df_comp_proc["pdays"].replace(-1, np.nan)

# Tratar 'unknown' como NaN en columnas categóricas
for col in ["job", "education", "contact", "poutcome"]:
    if col in df_comp_proc.columns:
        df_comp_proc[col] = df_comp_proc[col].replace("unknown", np.nan)

print("Datos preprocesados listos para predicción:")
display(df_comp_proc.head())
```

```python
# Realizar predicciones
predicciones_comp = pipeline_modelo_final.predict(df_comp_proc)

# Crear DataFrame de resultados
predicciones_df = pd.DataFrame({
    "deposit": predicciones_comp
})

# Guardar a CSV
predicciones_df.to_csv("predicciones.csv", index=False)

print("Predicciones guardadas exitosamente en 'predicciones.csv'")
display(predicciones_df.head())
```

Con esto, queda completado el flujo principal de modelado pedido en la práctica: selección de modelo, estimación del rendimiento futuro, entrenamiento del modelo final y generación de predicciones para los datos de competición.
